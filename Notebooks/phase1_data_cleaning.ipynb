Job cleaning & text creation

```python
import re

job_cols = [
    "job_id",
    "title",
    "description",
    "skills_desc",
    "formatted_experience_level",
    "formatted_work_type"
]

jobs = job_df[job_cols].copy()

# Drop very poor descriptions
jobs = jobs[jobs["description"].str.len() >= 100]

# Clean skills_desc
def clean_skills(text):
    if pd.isna(text):
        return ""
    text = re.sub(r"PI[a-zA-Z0-9\-]+", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

jobs["skills_desc"] = jobs["skills_desc"].apply(clean_skills)

# Combine text fields
jobs["clean_job_text"] = (
    jobs["title"].astype(str) + " " +
    jobs["description"].astype(str) + " " +
    jobs["skills_desc"].astype(str)
)

# Final text cleanup
jobs["clean_job_text"] = jobs["clean_job_text"].str.lower()
jobs["clean_job_text"] = jobs["clean_job_text"].str.replace(r"\s+", " ", regex=True)

jobs.shape
```

```python
jobs["formatted_experience_level"] = jobs["formatted_experience_level"].fillna("not specified")
```

```python
DATA_PROCESSED = os.path.join(PROJECT_ROOT, "Data", "processed")
jobs.to_csv(os.path.join(DATA_PROCESSED, "jobs_clean.csv"), index=False)
```
